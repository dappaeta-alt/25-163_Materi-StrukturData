```python
nama = ["Daffa","Tsaqib","Anasyah"]

# nama[1] = "Eta"

#menambah di akhir
nama.append("Evan")

#menambah di tempat yang ditentukan
nama.insert(1, "Yohanes")

#untuk menghapus pakai parameter indeks
nama.pop()

#menghapus dengan parameter nilai
nama.remove("Tsaqib")

print(nama[1])

for i in nama:
  print(i)
```
