# Sparse Matrix


```python
# ================================
# SPARSE MATRIX (DICTIONARY)
# ================================

# (a) Membuat Sparse Matrix
def buat_sparse(baris, kolom):
    sparse = {}
    print("Masukkan elemen matriks:")
    for i in range(baris):
        for j in range(kolom):
            nilai = int(input(f"Elemen [{i}][{j}] = "))
            if nilai != 0:
                sparse[(i, j)] = nilai
    return sparse


# (b) Menampilkan Sparse Matrix
def tampilkan_sparse(sparse, baris, kolom):
    for i in range(baris):
        for j in range(kolom):
            if (i, j) in sparse:
                print(sparse[(i, j)], end=" ")
            else:
                print(0, end=" ")
        print()


# (c) Penjumlahan Sparse Matrix
def tambah_sparse(A, B):
    hasil = {}

    # salin A
    for key in A:
        hasil[key] = A[key]

    # tambah B
    for key in B:
        if key in hasil:
            hasil[key] = hasil[key] + B[key]
            if hasil[key] == 0:
                del hasil[key]
        else:
            hasil[key] = B[key]

    return hasil


# (d) Perkalian Sparse Matrix
def kali_sparse(A, B, baris_A, kolom_A, kolom_B):
    hasil = {}

    for (i, k) in A:
        for j in range(kolom_B):
            if (k, j) in B:
                if (i, j) in hasil:
                    hasil[(i, j)] = hasil[(i, j)] + A[(i, k)] * B[(k, j)]
                else:
                    hasil[(i, j)] = A[(i, k)] * B[(k, j)]

    return hasil


# ================================
# PROGRAM UTAMA
# ================================

baris_A, kolom_A = 2, 2
baris_B, kolom_B = 2, 2

print("Matrix A")
A = buat_sparse(baris_A, kolom_A)

print("\nMatrix B")
B = buat_sparse(baris_B, kolom_B)

print("\nA:")
tampilkan_sparse(A, baris_A, kolom_A)

print("\nB:")
tampilkan_sparse(B, baris_B, kolom_B)

# Penjumlahan
C = tambah_sparse(A, B)
print("\nA + B:")
tampilkan_sparse(C, baris_A, kolom_A)

# Perkalian
D = kali_sparse(A, B, baris_A, kolom_A, kolom_B)
print("\nA x B:")
tampilkan_sparse(D, baris_A, kolom_B)
```
