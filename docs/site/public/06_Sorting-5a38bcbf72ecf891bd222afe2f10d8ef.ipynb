## Selection Sort


```python
# selection sort
def selectionSort(listData):
    print('Algoritma Selection Sort konvensional')
    print('Data Awal=',listData)
    for outIter in range(len(listData)-1):
        minIndex=outIter
        for i in range(outIter+1,len(listData)):
            if listData[i]<listData[minIndex]:
                minIndex=i
                
        temp=listData[outIter]
        listData[outIter]=listData[minIndex]
        listData[minIndex]=temp
        print('Iterasi ke-',outIter+1,':',listData)
    print('Data Urut=',listData)

a=[9,4,7,1]
selectionSort(a)
```

```python
def insertion_sort(arr):
    n = len(arr)
    
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        shifting = 0
        
        print(f"\nIterasi ke-{i}")
        print(f"Key = {key}")
        
        # Proses shifting
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
            shifting += 1
        
        # Menempatkan key
        arr[j + 1] = key
        
        print(f"Posisi akhir key = index {j+1}")
        print(f"Jumlah shifting = {shifting}")
        print(f"Array sekarang = {arr}")

# Contoh penggunaan
data = [7, 4, 5, 2]
print("Array awal:", data)

insertion_sort(data)

print("\nArray akhir:", data)
```

## Sorting Lanjutan (TPP 4) - Modifikasi & Perbandingan


# tpp 4

```python
# BUBBLE SORT DUA ARAH (MODIFIKASI)
def bubble_dua_arah(arr):
    data = arr[:]
    left = 0
    right = len(data) - 1
    iterasi = 1
    total_swap = 0

    while left < right:
        swapped = False

        if iterasi % 2 == 1:
            print("\nITERASI KE-", iterasi, "(GANJIL kiri ke kanan)")
            for i in range(left, right):
                print("(", data[i], "vs", data[i+1], ")", end=" ")

                if data[i] > data[i+1]:
                    data[i], data[i+1] = data[i+1], data[i]
                    total_swap += 1
                    swapped = True
                    print("swap", data)
                else:
                    print("tidak swap")

            right -= 1

        else:
            print("\nITERASI KE-", iterasi, "(GENAP kanan ke kiri)")
            for i in range(right, left, -1):
                print("(", data[i], "vs", data[i-1], ")", end=" ")

                if data[i] < data[i-1]:
                    data[i], data[i-1] = data[i-1], data[i]
                    total_swap += 1
                    swapped = True
                    print("swap", data)
                else:
                    print("tidak swap")

            left += 1

        print("Hasil Iterasi", iterasi, ":", data)

        if not swapped:
            print("Early stopping aktif berhenti")
            break

        iterasi += 1

    print("Total iterasi:", iterasi)
    print("Total swap:", total_swap)
    return data

# BUBBLE SORT BIASA
def bubble_biasa(arr):
    data = arr[:]
    n = len(data)
    iterasi = 0
    total_swap = 0

    for i in range(n):
        iterasi += 1
        swapped = False

        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                total_swap += 1
                swapped = True

        if not swapped:
            break

    print("\nBubble Biasa")
    print("Total iterasi:", iterasi)
    print("Total swap:", total_swap)
    return data

# SELECTION SORT MIN MAX
def selection_min_max(arr):
    data = arr[:]
    left = 0
    right = len(data) - 1
    iterasi = 1
    total_swap = 0

    while left <= right:
        print("\nITERASI KE-", iterasi)

        if iterasi % 2 == 1:
            min_idx = left
            for i in range(left, right + 1):
                if data[i] < data[min_idx]:
                    min_idx = i

            if min_idx != left:
                data[left], data[min_idx] = data[min_idx], data[left]
                total_swap += 1

            print("Cari MIN:", data)
            left += 1

        else:
            max_idx = left
            for i in range(left, right + 1):
                if data[i] > data[max_idx]:
                    max_idx = i

            if max_idx != right:
                data[right], data[max_idx] = data[max_idx], data[right]
                total_swap += 1

            print("Cari MAX:", data)
            right -= 1

        iterasi += 1

    print("Total iterasi:", iterasi - 1)
    print("Total swap:", total_swap)
    return data

# SELECTION SORT BIASA
def selection_biasa(arr):
    data = arr[:]
    n = len(data)
    iterasi = 0
    total_swap = 0

    for i in range(n):
        iterasi += 1
        min_idx = i

        for j in range(i + 1, n):
            if data[j] < data[min_idx]:
                min_idx = j

        if min_idx != i:
            data[i], data[min_idx] = data[min_idx], data[i]
            total_swap += 1

    print("\nSelection Biasa")
    print("Total iterasi:", iterasi)
    print("Total swap:", total_swap)
    return data

# INSERTION SORT MODIFIKASI
def insertion_modifikasi(arr):
    data = arr[:]
    iterasi = 0
    shifting = 0

    for i in range(1, len(data)):
        iterasi += 1
        key = data[i]

        if key >= data[i - 1]:
            continue

        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            shifting += 1
            j -= 1

        data[j + 1] = key

    print("\nInsertion Modifikasi")
    print("Total iterasi:", iterasi)
    print("Total shifting:", shifting)
    return data

# INSERTION SORT BIASA
def insertion_biasa(arr):
    data = arr[:]
    iterasi = 0
    shifting = 0

    for i in range(1, len(data)):
        iterasi += 1
        key = data[i]
        j = i - 1

        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            shifting += 1
            j -= 1

        data[j + 1] = key

    print("\nInsertion Biasa")
    print("Total iterasi:", iterasi)
    print("Total shifting:", shifting)
    return data


# =========================================
# MAIN
# =========================================
data1 = [50, 20, 40, 10, 30, 60]
data2 = [30, 10, 25, 5, 20, 15]
data3 = [10, 20, 30, 40, 50, 60]

print("===== BUBBLE DUA ARAH =====")
bubble_dua_arah(data1)

print("\n===== BUBBLE BIASA =====")
bubble_biasa(data1)

print("\n===== SELECTION MIN MAX =====")
selection_min_max(data2)

print("\n===== SELECTION BIASA =====")
selection_biasa(data2)

print("\n===== INSERTION MODIFIKASI =====")
insertion_modifikasi(data3)

print("\n===== INSERTION BIASA =====")
insertion_biasa(data3)
```
