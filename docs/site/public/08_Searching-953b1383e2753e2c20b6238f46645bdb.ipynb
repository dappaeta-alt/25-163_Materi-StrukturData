# Searching

## Linear Search


```python
# linear search (squencial search)
def linearSearch(arr, targetVal):
    for i in range(len(arr)):
        if arr[i] == targetVal:
            return i
    return -1

arr = [3, 7, 2, 9, 5]
targetVal = 9

result = linearSearch(arr, targetVal)

if result != -1:
    print("Value",targetVal,"found at index",result)
else:
    print("Value",targetVal,"not found")
```

```python
def seqSearch(listData, data):
    ind = 0
    found = False
    
    while ind < len(listData) and not found:
        if listData[ind] == data:
            found = True
        else:
            ind = ind + 1
    
    if found:
        return (f"ada di index = {ind}")
    else:
        return "Data tidak ada"


a = [12, 5, 9, 8, 1, 10, 26]

print(seqSearch(a, 1))   # output: 4
print(seqSearch(a, 7))   # output: Data tidak ada
```

## Binary Search


```python
# binary search
def binarySearch(alist, item):
    first = 0
    last = len(alist)-1
    found = False
    while first<=last and not found:
        midpoint = (first + last)//2
        if alist[midpoint] == item:
            found = True
        else:
            if item < alist[midpoint]:
                last = midpoint-1
            else:
                first = midpoint+1
    return found
    
testlist = [0, 1, 2, 8, 13, 17, 19, 32, 42,]
print(binarySearch(testlist, 3))
```

## Perbandingan Linear vs Binary Search


```python
import random

# =========================
# Linear Search
# =========================
def linear_search(data, target):
    iterasi = 0
    for i in range(len(data)):
        iterasi += 1
        if data[i] == target:
            return iterasi
    return iterasi  # tetap return jumlah iterasi walau tidak ditemukan


# =========================
# Binary Search
# =========================
def binary_search(data, target):
    left = 0
    right = len(data) - 1
    iterasi = 0

    while left <= right:
        iterasi += 1
        mid = (left + right) // 2

        if data[mid] == target:
            return iterasi
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return iterasi


# =========================
# Data Random (100 elemen)
# =========================
data = random.sample(range(1, 1000), 100)
data_sorted = sorted(data)

# Target sesuai kondisi
target_awal = data[0]
target_tengah = data[len(data)//2]
target_akhir = data[-1]
target_tidak_ada = 1001  # di luar range


# =========================
# Fungsi Pengujian
# =========================
def uji(nama, target):
    print(f"\n=== {nama} (target = {target}) ===")

    it_lin_acak = linear_search(data, target)
    it_lin_urut = linear_search(data_sorted, target)
    it_bin = binary_search(data_sorted, target)

    print("Linear (tidak terurut):", it_lin_acak, "iterasi")
    print("Linear (terurut):     ", it_lin_urut, "iterasi")
    print("Binary Search:        ", it_bin, "iterasi")


# =========================
# Jalankan semua kondisi
# =========================
uji("Data di Awal", target_awal)
uji("Data di Tengah", target_tengah)
uji("Data di Akhir", target_akhir)
uji("Data Tidak Ada", target_tidak_ada)
```

```python
def linear_search_modifikasi(data, target):
    index_list = []

    for i in range(len(data)):
        if data[i] == target:
            index_list.append(i)

    if len(index_list) == 0:
        print("Data tidak ditemukan")
    else:
        print("Index:", index_list)
        print("Jumlah:", len(index_list))
        print("First:", index_list[0])
        print("Last:", index_list[-1])


# Data
data = [12, 5, 9, 8, 1, 10, 26, 5, 9, 5]

# Cari nilai 5
linear_search_modifikasi(data, 5)
```
