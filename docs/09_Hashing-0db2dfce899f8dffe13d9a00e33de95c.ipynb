```python
def remainderFunction (data,num):
    return (data%num)

def createHashTable(num):
    temp=[]
    for i in range(num):
        temp.append('none')
    return(temp)

def putData(data,table):
    for i in range(len(data)):
        ind=remainderFunction(data[i],len(table))
        table[ind]=data[i]
    return(table)

def searchHash(data,table):
    hashVal=remainderFunction(data,len(table))
    if data==table[hashVal]:
        return True
    else:
        return False

a=[54, 26, 93, 17, 77, 31]
hashTable=createHashTable(11)
print(hashTable)
```

```python
def midSqFunction(data,table_size):
    # Mengkuadratkan data dan mengubahnya menjadi string
    square = str(data ** 2)

    # menentukan titik tengah
    midInd = len(square) // 2

    # mengambil bagian tengah string
    if len(square) < 3:
        # jika hasil kuadrat kurang dari 3 digit, ambil seluruhnya
        midVal = int(square)
    elif len(square) % 2 == 0:
        # Jika jumlah digit genp, ambil 2 digit tengah
        midVal = int(square[midInd-1:midInd+1])
    else:
        # jika jumlah digit ganjil, ambil 2 digit tepat tengah
        midVal = int(square[midInd])

    return midVal % table_size
```

## Chaining


```python
def createChainingTable(num):
    temp = []
    for i in range(num):
        temp.append([])
    return temp

def hashFunction(data, num):
    return data%num

# create (insert)
def insertChaining(data,table):
    ind = hashFunction(data,len(table))
    table[ind].append(data)
    print(f"Insert: {data} dimasukkan ke slot {ind}")
    return table

# read
def searchChaining(data, table):
    ind = hashFunction(data, len(table))

    n = len(table[ind])
    for i in range(n):
        if data == table[ind][i]:
            print(f"Search: {data} ditemukan di slot {ind}")
            return True
        print(f"Search: {data} tidak ditemukan")
        return False

# Delete (Search dan Delete)
def deleteChaining(data, table):
    ind = hashFunction(data, len(table))

    n = len(table[ind])
    for i in range(n):
        if data == table[ind][i]:
            table[ind].pop(i)
            print(f"Delete: {data} berhasil dihapus dari slot {ind}")
            return True
        print(f"Delete: {data} gagal dihapus (data tidak ditemukan)")
        return False

# Update (search dan update)
def updateChaining(dataLama, dataBaru, table):
    print(f"Update {dataLama} menjadi {dataBaru}")

    if deleteChaining(dataLama, table):
        insertChaining(dataBaru, table)
        print(f"Update: {dataLama} diganti menjadi {dataBaru}")
    print(f"Update Gagal {dataLama} tidak ditemukan di tabel")
```

```python
print("--- TEST CHAINING ---")
# Create
tabel_chain = createChainingTable(5)
insertChaining(10, tabel_chain)
insertChaining(15, tabel_chain)
insertChaining(21, tabel_chain)
print("Tabel awal:", tabel_chain)

# Read
searchChaining(15, tabel_chain)
searchChaining(99, tabel_chain)

# Update
updateChaining(10, 99, tabel_chain)

# Delete
deleteChaining(15, tabel_chain)
print("Tabel akhir:", tabel_chain)
print("\n")
```

## Linear Probing (CRUD)


```python
def createHashTable(num):
    temp = []
    for i in range(num):
        temp.append("none")
    return temp

# create
def insertLinear(data, table):
    n = len(table)
    startInd = hashFunction(data, n)
    ind = startInd

    while table[ind] != "none" and table[ind] != "Deleted":
        ind = (ind+1) % n
        if ind == startInd:
            print("Tabel Penuh!")
            return False
    table[ind] = data
    return True

# read
def searchLinear(data, table):
    n = len(table)
    startInd = hashFunction(data, n)
    ind = startInd

    while table[ind] != "none":
        if table[ind] == data:
            return ind
        ind = (ind+1)%n
        if ind == startInd:
            break
    return -1

# delete
def deleteLinear(data, table):
    ind = searchLinear(data, table)
    if ind != -1:
        table[ind] = "Deleted"
        return True
    return False

# Update
def updateLinear(dataLama, dataBaru, table):
    if deleteLinear(dataLama, table):
        insertLinear(dataBaru, table)
        print(f"Update {dataLama} diubah menjadi {dataBaru}")
        return True
    return False
```

```python
print("--- TEST LINEAR PROBING ---")
# Create
tabel_chain = createHashTable(5)
insertLinear(10, tabel_chain)
insertLinear(15, tabel_chain)
insertLinear(21, tabel_chain)
print("Tabel awal:", tabel_chain)

# Read
print(searchLinear(15, tabel_chain))
print(searchLinear(99, tabel_chain))

# Update
updateLinear(10, 99, tabel_chain)

# Delete
deleteLinear(15, tabel_chain)
print("Tabel akhir:", tabel_chain)
print("\n")
```

## Quadratic Probing


```python
# create
def insertQuadratic(data, table):
    n = len(table)
    startInd = hashFunction(data, n)

    for i in range(n):
        ind = (startInd + i**2) % n
        if table[ind] == "none" or table[ind] == "Deleted":
            table[ind] = data
            return True

    print("Tabel penuh")
    return False

# read
def searchQuadratic(data, table):
    n = len(table)
    startInd = hashFunction(data,n)

    for i in range(n):
        ind = (startInd + i**2) % n
        if table[ind] == data:
            return ind
        if table[ind] == "none":
            break
    return -1

# delete
def deleteQuadratic(data, table):
    ind = searchQuadratic(data, table)
    if ind != -1:
        table[ind] = "Deleted"
        return True
    return False

# Update
def updateQuadratic(dataLama, dataBaru, table):
    if deleteQuadratic(dataLama, table):
        insertQuadratic(dataBaru, table)
        print(f"Update: {dataLama} diubah menjadi {dataBaru}")
        return True
    print("Update gagal, data lama tidak ditemukan")
    return False
```

```python
print("--- TEST QUADRATIC PROBING ---")
# Create
tabel_chain = createHashTable(5)
insertQuadratic(10, tabel_chain)
insertQuadratic(15, tabel_chain)
insertQuadratic(21, tabel_chain)
print("Tabel awal:", tabel_chain)

# Read
print(searchQuadratic(15, tabel_chain))
print(searchQuadratic(99, tabel_chain))

# Update
updateQuadratic(10, 99, tabel_chain)

# Delete
deleteQuadratic(15, tabel_chain)
print("Tabel akhir:", tabel_chain)
print("\n")
```

## Hash Class (OOP) - Chaining, Linear, Quadratic, MidSquare


```python
class HashChaining:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return key % self.size

    # CREATE
    def insert(self, key):
        index = self.hash_function(key)
        if key not in self.table[index]:
            self.table[index].append(key)

    # READ
    def search(self, key):
        index = self.hash_function(key)
        return key in self.table[index]

    # UPDATE
    def update(self, old, new):
        index = self.hash_function(old)
        if old in self.table[index]:
            self.table[index].remove(old)
            self.insert(new)

    # DELETE
    def delete(self, key):
        index = self.hash_function(key)
        if key in self.table[index]:
            self.table[index].remove(key)

    def display(self):
        print(self.table)
```

```python
class LinearProbing:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        start = index

        while self.table[index] is not None:
            index = (index + 1) % self.size
            if index == start:
                print("Tabel penuh")
                return
        self.table[index] = key

    def search(self, key):
        index = self.hash_function(key)
        start = index

        while self.table[index] is not None:
            if self.table[index] == key:
                return True
            index = (index + 1) % self.size
            if index == start:
                return False
        return False

    def update(self, old, new):
        if self.delete(old):
            self.insert(new)

    def delete(self, key):
        index = self.hash_function(key)
        start = index

        while self.table[index] is not None:
            if self.table[index] == key:
                self.table[index] = None
                return True
            index = (index + 1) % self.size
            if index == start:
                return False
        return False

    def display(self):
        print(self.table)
```

```python
class QuadraticProbing:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        i = 1

        while self.table[index] is not None:
            index = (index + i*i) % self.size
            i += 1
            if i == self.size:
                print("Tabel penuh")
                return
        self.table[index] = key

    def search(self, key):
        index = self.hash_function(key)
        i = 1

        while self.table[index] is not None:
            if self.table[index] == key:
                return True
            index = (index + i*i) % self.size
            i += 1
            if i == self.size:
                return False
        return False

    def update(self, old, new):
        if self.delete(old):
            self.insert(new)

    def delete(self, key):
        index = self.hash_function(key)
        i = 1

        while self.table[index] is not None:
            if self.table[index] == key:
                self.table[index] = None
                return True
            index = (index + i*i) % self.size
            i += 1
            if i == self.size:
                return False
        return False

    def display(self):
        print(self.table)
```

```python
class MidSquareHash:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def midsquare_hash(self, key):
        square = key * key
        mid = len(str(square)) // 2
        mid_digits = int(str(square)[mid-1:mid+1])
        return mid_digits % self.size

    def insert(self, key):
        index = self.midsquare_hash(key)
        while self.table[index] is not None:
            index = (index + 1) % self.size
        self.table[index] = key

    def search(self, key):
        index = self.midsquare_hash(key)
        start = index

        while self.table[index] is not None:
            if self.table[index] == key:
                return True
            index = (index + 1) % self.size
            if index == start:
                return False
        return False

    def update(self, old, new):
        if self.delete(old):
            self.insert(new)

    def delete(self, key):
        index = self.midsquare_hash(key)
        start = index

        while self.table[index] is not None:
            if self.table[index] == key:
                self.table[index] = None
                return True
            index = (index + 1) % self.size
            if index == start:
                return False
        return False

    def display(self):
        print(self.table)
```
