# Tugas Praktikum - Struktur Data


```python
# tugas praktikum 1
def indeks_maksimal(data):
    n = len(data)
    maks = (data[0]-1) * (data[1]-1)
    idx_i, idx_j = 0, 1

    for i in range(n):
        for j in range(i+1, n):
            nilai = (data[i]-1) * (data[j]-1)
            if nilai > maks:
                maks = nilai
                idx_i, idx_j = i, j

    return [idx_i, idx_j]


# Contoh penggunaan
data1 = [1,5,9,10,8]
data2 = [8,1,7,9,9,10,12,18,2,3]
data3 = [31,2,10,6,12,11,10,6,5,5]

print("input :", data1)
print("output:", indeks_maksimal(data1))

print("\ninput :", data2)
print("output:", indeks_maksimal(data2))

print("\ninput :", data3)
print("output:", indeks_maksimal(data3))
```

```python
# tugas praktikum 2
def cek_palindrome(data):
    # jika angka negatif langsung False
    if data < 0:
        return False

    # ubah angka menjadi string
    angka = str(data)

    # bandingkan dengan kebalikannya
    if angka == angka[::-1]:
        return True
    else:
        return False


# contoh penggunaan
print("input :", 125521)
print("output:", cek_palindrome(125521))

print("\ninput :", -125521)
print("output:", cek_palindrome(-125521))
```

```python
# tugas praktikum 3
def angka_terbanyak(data):
    frekuensi = {}

    # menghitung jumlah kemunculan setiap angka
    for angka in data:
        if angka in frekuensi:
            frekuensi[angka] += 1
        else:
            frekuensi[angka] = 1

    # mencari angka dengan frekuensi terbesar
    maks = 0
    hasil = None

    for angka, jumlah in frekuensi.items():
        if jumlah > maks:
            maks = jumlah
            hasil = angka

    return hasil


# contoh penggunaan
data1 = [4,1,4,6]
data2 = [1,5,5,8,9,10,12]
data3 = [2,2,1,1,1,2,2]

print("input :", data1)
print("output:", angka_terbanyak(data1))

print("\ninput :", data2)
print("output:", angka_terbanyak(data2))

print("\ninput :", data3)
print("output:", angka_terbanyak(data3))
```

```python
# tugas praktikum 4
# fungsi membuat list
def createList(n):
    data = []
    for i in range(n):
        nilai = int(input(f"Masukkan data ke-{i+1}: "))
        data.append(nilai)
    return data


# fungsi menghitung rata-rata (persamaan 1)
def hitung_rata(data):
    n = len(data)
    total = sum(data)
    y = total / n
    return y


# fungsi menghitung varians (persamaan 2)
def hitung_varians(data, y):
    n = len(data)
    jumlah = 0

    for x in data:
        jumlah += (x - y) ** 2

    z = jumlah / (n - 1)
    return z


# program utama
n = int(input("Masukkan jumlah data: "))

data = createList(n)

print("List data:", data)

y = hitung_rata(data)
print("Rata-rata (y) =", y)

z = hitung_varians(data, y)
print("Varians (z) =", z)
```
