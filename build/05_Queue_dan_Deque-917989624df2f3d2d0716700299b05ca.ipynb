## Challenge - Hot Potato


```python
def queue():
    q = []
    return q

def enqueue(q, data):
    q.append(data)
    return q

def dequeue(q):
    data = q.pop(0)
    return data

def isEmpty(q):
    return (q == [])

def hot_potato(n, k):
    q = queue()

    # isi pemain
    for i in range(1, n + 1):
        enqueue(q, i)

    putaran = 1

    while len(q) > 1:
        print(f"\nPutaran {putaran}:")

        # oper kentang
        for i in range(1, k + 1):
            pemain = dequeue(q)
            print(f"  Potato pada orang {pemain}")
            enqueue(q, pemain)

        # orang terakhir kena potato → keluar
        keluar = dequeue(q)
        print(f"  ➤ Orang {keluar} keluar (terakhir pegang potato)")

        putaran += 1

    # pemenang
    print(f"\nPemenang: Orang {q[0]}")
    return q[0]

hot_potato(5,2)
```

```python
def queue():
    q = []
    return q

def enqueue(q, data):
    q.append(data)
    return q

def dequeue(q):
    data = q.pop(0)
    return data

def isEmpty(q):
    return (q == [])

def hot_potato(n, k):
    q = queue()

    # isi pemain
    for i in range(1, n + 1):
        enqueue(q, i)

    putaran = 1

    while len(q) > 1:
        print(f"\nPutaran {putaran}:")
        print(f"Sisa pemain: {q}")

        # oper kentang sampai k
        for i in range(1, k + 1):
            orang = dequeue(q)
            print(f"  Potato pada orang {orang}")

            # kalau belum ke-k, masukkan lagi
            if i != k:
                enqueue(q, orang)

        # orang ke-k keluar
        print(f"  ➤ Orang {orang} keluar")

        putaran += 1

    print(f"\nPemenang: Orang {q[0]}")
    return q[0]

hot_potato(5, 2)
```
