# Konversi Ekspresi: Infix, Postfix, Prefix


```python
# Fungsi menentukan prioritas operator
def prioritas(op):
    if op in ['+', '-']:
        return 1
    elif op in ['*', '/']:
        return 2
    elif op == '^':
        return 3
    return 0


# Fungsi konversi infix ke postfix
def infix_ke_postfix(ekspresi):
    stack = []
    hasil = ""

    for char in ekspresi:
        # operand
        if char.isalnum():
            hasil += char
        
        # kurung buka
        elif char == '(':
            stack.append(char)
        
        # kurung tutup
        elif char == ')':
            while stack and stack[-1] != '(':
                hasil += stack.pop()
            stack.pop()
        
        # operator
        else:
            while stack and prioritas(char) <= prioritas(stack[-1]):
                hasil += stack.pop()
            stack.append(char)

    while stack:
        hasil += stack.pop()

    return hasil


# Fungsi konversi infix ke prefix
def infix_ke_prefix(ekspresi):
    # balik ekspresi
    ekspresi = ekspresi[::-1]

    # tukar kurung
    ekspresi_baru = ""
    for c in ekspresi:
        if c == '(':
            ekspresi_baru += ')'
        elif c == ')':
            ekspresi_baru += '('
        else:
            ekspresi_baru += c

    # ubah ke postfix
    postfix = infix_ke_postfix(ekspresi_baru)

    # balik hasil postfix
    prefix = postfix[::-1]

    return prefix


# Program utama
infix = input("Masukkan ekspresi infix: ")

postfix = infix_ke_postfix(infix)
prefix = infix_ke_prefix(infix)

print("Postfix :", postfix)
print("Prefix  :", prefix)
```

```python
# Program Postfix ke Infix

postfix = input("Masukkan ekspresi postfix: ")

stack = []

for i in postfix:
    if i.isalnum():          # jika operand
        stack.append(i)
    else:                    # jika operator
        op2 = stack.pop()
        op1 = stack.pop()
        ekspresi = "(" + op1 + i + op2 + ")"
        stack.append(ekspresi)

hasil = stack.pop()

print("Hasil Infix:", hasil)
```

```python
# Program Prefix ke Infix

prefix = input("Masukkan ekspresi prefix: ")

stack = []

# Membaca dari kanan ke kiri
for i in reversed(prefix):
    if i.isalnum():           # jika operand
        stack.append(i)
    else:                     # jika operator
        op1 = stack.pop()
        # op2 = stack.pop()
        ekspresi = "(" + op1 + i + ")"
        stack.append(ekspresi)

hasil = stack.pop()

print("Hasil Infix:", hasil)
```

```python
# Program Prefix ke Postfix

prefix = input("Masukkan ekspresi prefix: ")

stack = []

# membaca dari kanan ke kiri
for i in reversed(prefix):
    if i.isalnum():      # jika operand
        stack.append(i)
    else:                # jika operator
        op1 = stack.pop()
        op2 = stack.pop()
        ekspresi = op1 + op2 + i
        stack.append(ekspresi)

hasil = stack.pop()

print("Hasil Postfix:", hasil)
```

```python
# Program Postfix ke Prefix

postfix = input("Masukkan ekspresi postfix: ")

stack = []

for i in postfix:
    if i.isalnum():      # jika operand
        stack.append(i)
    else:                # jika operator
        op2 = stack.pop()
        op1 = stack.pop()
        ekspresi = i + op1 + op2
        stack.append(ekspresi)

hasil = stack.pop()

print("Hasil Prefix:", hasil)
```
